**Data Pre-Profiling on Sodium**

Sodium has integer values but it's showing object data type.

We conclude that some issue occur in sodium column so that We have to rectify that one.

Convert the object data type of Sodium into int.

```python
df_mcdmenu["Sodium"].value_counts()
```

**Here all values displayed in continuous variable means it has in incorrect data type.some of the values filled with underscore. First we have to remove the underscore and fill the values with median.**

**Data Pre_Processing**  : To fix the issue by doing some modifications on dataset..

```python
df_mcdmenu["Sodium"]=='-'
```

```python
df_mcdmenu.loc[df_mcdmenu["Sodium"]=='-',"Category"].nunique()  # it displays the count value
```

```python
df_mcdmenu.loc[df_mcdmenu["Sodium"]=='-',"Category"].unique()  # it displays the names of the category
```

**In the category column the Cold Beverages and Hot Beverages have underscore.**

```python
df_mcdmenu.loc[df_mcdmenu['Category'] == 'Cold Beverages', 'Sodium'] = df_mcdmenu.loc[df_mcdmenu['Category'] == 'Cold Beverages', 'Sodium'].replace('-',0)
```

**Here we filled zero in place of underscore for Cold Beverages in Category column**.

```python
df_mcdmenu["Sodium"].value_counts()
```

```python
df_mcdmenu.loc[df_mcdmenu['Category'] == 'Hot Beverages'].value_counts()
```

**In Hot Beverages, some of the records have value 20 in sodium column.So we have to fill the remaining records with 20.**

```python
df_mcdmenu.loc[df_mcdmenu['Category'] == 'Hot Beverages', 'Sodium'] = df_mcdmenu.loc[df_mcdmenu['Category'] == 'Hot Beverages', 'Sodium'].replace('-',20)
```

```python
df_mcdmenu.loc[df_mcdmenu['Category'] == 'Hot Beverages'].value_counts()
```

Here we filled all the values with 20.

```python
df_mcdmenu["Sodium"] =df_mcdmenu["Sodium"].astype(int)
```

**Correlation in Python** :
Correlation summarizes the strength and direction of the linear (straight-line) association between two quantitative variables. Denoted by r, it takes values between -1 and +1. A positive value for r indicates a positive association, and a negative value for r indicates a negative association. The closer r is to 1 the closer the data points fall to a straight line, thus, the linear association is stronger. The closer r is to 0, making the linear association weaker.
